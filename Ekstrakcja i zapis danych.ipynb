```python
#Import bibliotek
import pandas as pd
import csv as csv
import numpy as np
from os import listdir
```

```python
#Zmienne konfiguracyjne
Dirpath_s_d = "C:\\Users\\Win10\\Desktop\\semestr2\\Metody analizy i ekspolaracji danych\\Dane\\s_d"
Dirpath_s_d_t = "C:\\Users\\Win10\\Desktop\\semestr2\\Metody analizy i ekspolaracji danych\\Dane\\s_d_t"
_SavePath = "C:\\Users\\Win10\\Desktop\\semestr2\\Metody analizy i ekspolaracji danych\\Dane\\Dane_zebrane.csv"
Template_path_s_d = "C:\\Users\\Win10\\Desktop\\semestr2\\Metody analizy i ekspolaracji danych\\Dane\\Templates\\template_s_d.csv"
Template_path_s_d_t = "C:\\Users\\Win10\\Desktop\\semestr2\\Metody analizy i ekspolaracji danych\\Dane\\Templates\\template_s_d_t.csv"
```

```python
#Wczytanie i połączenie wszystkich plików .csv z jednej ścieżki
filenames = listdir(Dirpath_s_d)
tabela = pd.read_csv(filepath_or_buffer = Template_path_s_d, delimiter = ";", encoding = 'latin1')
for filename in filenames:
   if filename.endswith(".csv"):
        tabela_temp = pd.read_csv(filepath_or_buffer = Dirpath_s_d + "\\" + filename, delimiter = ",", encoding='latin1',header=None)
        tabela_temp.columns = list(tabela.columns)
        tabela=pd.concat([tabela,tabela_temp],ignore_index=True)
```

```python
#Wczytanie i połączenie wszystkich plików .csv z jednej ścieżki
filenames = listdir(Dirpath_s_d_t)
tabela2 = pd.read_csv(filepath_or_buffer = Template_path_s_d_t, delimiter = ";", encoding = 'latin1')
for filename in filenames:
   if filename.endswith(".csv"):
        tabela_temp = pd.read_csv(filepath_or_buffer = Dirpath_s_d_t + "\\" + filename, delimiter = ",", encoding='latin1',header=None)
        tabela_temp.columns = list(tabela2.columns)
        tabela2=pd.concat([tabela2,tabela_temp],ignore_index=True)
```

```python
tabela2.drop(columns={'Kod stacji','Nazwa stacji','Rok','Miesiac','Dzien'},inplace=True)
tabela_final=pd.concat([tabela,tabela2],ignore_index=False,axis=1,verify_integrity=True)
tabela_final
```

```python
df = tabela_final[['Rok','Miesiac','Dzien']].copy()
df.rename(columns={'Rok': 'year','Miesiac': 'month', 'Dzien': 'day'}, inplace=True)
df = pd.to_datetime(df)
tabela_final['Data'] = df
#tabela_final.drop(columns={'Rok','Miesiac','Dzien'},inplace=True)
```

```python
tabela_final.drop_duplicates(subset=['Data'], inplace=True)
tabela_final.set_index('Data',inplace = True)
```

```python
tabela_final.to_csv(path_or_buf=_SavePath, sep=';')
tabela_final
```
